```python
# fetch weather data from Environment Canada & clean it 
weather = pd.read_html("http://climate.weather.gc.ca/climate_data/daily_data_e.html?StationID=27174&timeframe"
                       + "=2&StartYear=1840&EndYear=2019&Day=11&Year=2019&Month=6#")[0]
weather.drop(index=[30,31,32,33],axis=0,inplace=True)
weather.drop(axis=1,columns=["Total Rain Definition mm","Snow on Grnd Definition cm"],inplace=True)
weather.fillna(0,inplace=True)
weather
```

```python
weather[weather["Mean Temp Definition °C"]>20]
```

```python
weather.info()
```

```python
weather = weather.apply(pd.to_numeric)
weather.info()
```

```python
degree_sign = '\N{degree sign}'
print(degree_sign)
```
