Import numpy, pandas, matplotlib.pyplot and seaborn

Import the "fruit.csv" file into a DataFrame called fruit.<br>
Set the index to column 0.<br>
Display the fruit DataFrame.

```python
# Import the CSV file into a DataFrame with the specified index
fruit = pd.read_csv('fruit.csv', index_col=0)

# Display the fruit DataFrame
fruit
```

Use matplotlib plot function to plot x=Volume vs. y=Weight of ALL the data.<br>
Use marker style "bo".<br>
Label your axes and title the chart: Apples vs. Oranges.

```python
# Plotting Volume vs. Weight
plt.plot(fruit['Volume'], fruit['Weight'], 'bo', label='Data')

# Adding labels and title
plt.xlabel('Volume')
plt.ylabel('Weight')
plt.title('Apples vs. Oranges')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

Practice your pandas data selection and make two plots on the same canvas:<br>
1. Weight vs. Volume of ONLY Apples, use blue for the marker color.<br>
2. Weight vs. Volume of ONLY Oranges, use red for the marker color.<br>
Add labels, title and a legend.

```python
# Filter data for apples and oranges
apples = fruit[fruit['Type'] == 'Apple']
oranges = fruit[fruit['Type'] == 'Orange']

# Plot Weight vs. Volume for apples
plt.plot(apples['Volume'], apples['Weight'], 'bo', label='Apples')

# Plot Weight vs. Volume for oranges
plt.plot(oranges['Volume'], oranges['Weight'], 'ro', label='Oranges')

# Adding labels and title
plt.xlabel('Volume')
plt.ylabel('Weight')
plt.title('Weight vs. Volume of Apples and Oranges')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

Use seaborn to create a scatterplot of ALL the data.<br>
x=Volume<br>
y=Weight

```python
# Create scatterplot using Seaborn
sns.scatterplot(data=fruit, x='Volume', y='Weight')

# Adding labels and title
plt.xlabel('Volume')
plt.ylabel('Weight')
plt.title('Weight vs. Volume Scatterplot')

# Display the plot
plt.show()
```

Let us find the centroids of the clusters (the mean x- and y- values for each type of fruit). Calculate the mean values of volume and weight for the apples, and for the oranges. Store the volumes in a list called 'x'; store the weights in a list called 'y'. Some of the code has been done for you, but you need to add more to the cell below.

```python
apple = fruit[fruit['Type']=='Apple'][['Volume', 'Weight']].mean()
orange = fruit[fruit['Type']=='Orange'][['Volume', 'Weight']].mean()

# Store the mean volumes and weights in lists
x = [apple['Volume'], orange['Volume']]
y = [apple['Weight'], orange['Weight']]

# Display the mean values
print("x :", x)
print("y :", y)


```

Create a scatter plot of all the data. Use the hue option to color the data by type of fruit. Plot the x and y lists created above on top of this plot, using a cyan star of size '200'.

```python
# Create scatterplot of all the data colored by type of fruit
sns.scatterplot(data=fruit, x='Volume', y='Weight', hue='Type')

# Plot centroids on top of the scatterplot
plt.scatter(x, y, marker='*', color='cyan', s=200, label='Centroids')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

How balanced is the data? Use seaborn to produce barplot showing the count of apples and oranges.

```python
# Create a barplot showing the count of apples and oranges
sns.countplot(data=fruit, x='Type')

# Adding labels and title
plt.xlabel('Fruit Type')
plt.ylabel('Count')
plt.title('Count of Apples and Oranges')

# Display the plot
plt.show()
```

Use seaborn to create a jointplot of only the apple data.

```python
# Filter the 'fruit' DataFrame to select only the rows corresponding to apples
apples_data = fruit[fruit['Type'] == 'Apple']

# Create a jointplot of only the apple data
sns.jointplot(data=apples_data, x='Volume', y='Weight', kind='scatter')

# Display the plot
plt.show()
```

Use seaborn distplots to visualize the distributions of the weight and volume of the apples.

```python
# Create distplots for weight and volume of apples
plt.figure(figsize=(10, 5)) 
plt.subplot(1, 2, 1)
sns.histplot(apples_data['Weight'], kde=True, color='blue')
plt.title('Distribution of Weight for Apples')

plt.subplot(1, 2, 2)
sns.histplot(apples_data['Volume'], kde=True, color='green')
plt.title('Distribution of Volume for Apples')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
